# Publication topic trends in _Nature_

For my final project for Harvard's (2014) intro to computer science course, I wanted to get some practice using Python and building a web scraper. To do  achieve these goals, I decided to look at publication trends in _Nature_, which is one of the major journals scientific journals we look to for cutting edge science. Specifically, I wanted to see whether there are certain topics that are more likely to be published than others, and whether this changes over time  (e.g. biological sciences vs physical sciences). Has there been a recent surge in papers published on the biological sciences or has this been pretty consistent over the past several years? Are there any topics that are hardly represented by Nature? These findings will be informative both in terms of what the "hot" topics in science are and have been, and whether Nature may have some bias towards which kinds of stories it accepts for publication. Ultimately, I think it would be interesting to apply the same analysis to _Cell_ and _Science_, but I figured I _Nature_ would be a good start.

## Get Data

To obtain data, I am using BeautifulSoup to scrape _Nature's_ website. For most papers, _Nature_ assigns a number of subject terms that aim to capture the "themes" of the paper. You can see an example of this highlighted in pink below. The goal is to scrape all of the subject terms for every Letter and Article (types of papers) in _Nature_. This will allow me to assay whether there are any particular trends in which topics _Nature_ publishes.

<img src="Nature_subject_terms_example.png" alt="Drawing" style="width: 500px;"/>

```
# Get Nature issue index (from present to 2010, first year where subject terms were included)
url = 'http://www.nature.com/nature/archive/index.html?year=2014-2013-2012-2011-2010&remove-year=2009#2009'
source = urllib2.urlopen(url).read()

# get bs4 object
soup = bs4.BeautifulSoup(source)
```

```
# Get all links on page
link_list = [l.get('href') for l in soup.findAll('a')]

# Let's filter those objects to get only relevant links (exclude supplementary info link as well)
sorted_links = []

# write a loop to filter the links
for l in link_list:
    if l is not None and l[:16] == '/nature/journal/' and l[-16:] !=  '_supp/index.html':
        sorted_links.append(l)
```

Turns out that Nature only started including subject ID's during issue 7292, so lets drop issues < 7292 (last 15)

```
# Get number of elements in links list
num_links = len(sorted_links)

# Subtract 15
rev_num_links = num_links - 15

# take only the number of valid links (which includes none that are < 7292)
valid_issue_links = sorted_links[:rev_num_links]

# get all links except for the first one
valid_issue_links_minus_first = valid_issue_links[-235:]

# And, unfortunately, it looks like the site layout changed for issues 7313 and below, so I have a modified for loop for these
valid_issue_links_group_7 = valid_issue_links_minus_first[213:235]

# all valid links minus the first and < 7313
valid_issue_links_group_8 = valid_issue_links_minus_first[0:213]
```

```
# Get article names/links per issue (Only selecting Articles and Letters (Primary Literature))
# Run through the loop with one issue at first to get the initial df that we will build on in the 
# for loop below
url = 'http://nature.com/nature/journal/v515/n7527/index.html'

# get bs4 object
source2 = urllib2.urlopen(url).read()
soup2 = bs4.BeautifulSoup(source2)

# Get relevant tag containing **Articles** info
article_section = soup2.find("div", {"id": "af"})
article_section = article_section.find("div", {"class":"sub-inner-content"})

# Get title and link to paper, get all 'a'
article_a = article_section.findAll('a')
    
# Get relevant tag containing **Letters** info
letter_section = soup2.find("div", {"id": "lt"})
letter_section = letter_section.find("div", {"class":"sub-inner-content"})

# Get title and link to paper, get all 'a'
letter_a = letter_section.findAll('a')

# Get link for articles and letters
a_link_list = pd.DataFrame(l.get('href') for l in article_a)
a_link_list = a_link_list.rename(columns = {0:'Paper_link'})

l_link_list = pd.DataFrame(l.get('href') for l in letter_a)
l_link_list = l_link_list.rename(columns = {0:'Paper_link'})

# Get Paper name in letters and articles
a_names = [l.contents for l in article_a]
l_names = [l.contents for l in letter_a]
a_l_names = [a_names, l_names]

# Get rid of all the crap in the name
for names in a_l_names:
    for x in xrange(len(names)):
        names[x] = str(names[x])
        names[x] = names[x].replace(',','')
        names[x] = names[x].replace('<i>','')
        names[x] = names[x].replace('</i>','')
        names[x] = names[x].replace('<span class="mb">','')
        names[x] = names[x].replace('</span>','')
        names[x] = names[x].replace('<sup>','')
        names[x] = names[x].replace('</sup>','')
        names[x] = names[x].replace('\u2032','')
        names[x] = names[x].replace('\u2019','')
        names[x] = names[x].replace('<sub>','')
        names[x] = names[x].replace('</sub>','')
        names[x] = names[x].replace("[u'",'')
        names[x] = names[x].replace("'", '')
        names[x] = names[x].replace("]", '')
        names[x] = names[x].replace("[", '')
        names[x] = names[x].replace("u'", '')
        names[x] = names[x].replace("\\u2013", ' ')
        
# Convert to dataframe
l_names = pd.DataFrame(l_names)
l_names = l_names.rename(columns = {0:'Paper_name'})

a_names = pd.DataFrame(a_names)
a_names = a_names.rename(columns = {0:'Paper_name'})

# join articles and letters info
article_df = pd.concat([a_names, a_link_list], axis = 1)
article_df['Paper_type'] = 'article'

letter_df = pd.concat([l_names, l_link_list], axis = 1)
letter_df['Paper_type'] = 'letter'
    
# join Letter and Articles dfs
all_papers = pd.concat([letter_df, article_df], axis = 0, ignore_index = True)
    
# Add paper volume and issue info
all_papers['volume'] = url[34:37]
all_papers['issue'] = url[39:43]
    
# Add publication date
pub_date_html = soup2.find("span", {"class": "more"})
all_papers['pub_date'] = pub_date_html.contents[0]

# Get rid of those pesky News and Views articles and references to other Articles/Letters
sorted_all_papers = all_papers[~all_papers['Paper_name'].str.contains('News & Views')]
sorted_all_papers = sorted_all_papers[~sorted_all_papers['Paper_name'].str.contains('et al.')]
```

```
# Add in papers from idexes with changed html formats
for url in valid_issue_links_group_7: 
    url2 = 'http://www.nature.com' + url

    source3 = urllib2.urlopen(url2).read()

    # get bs4 object
    soup3 = bs4.BeautifulSoup(source3)

    # Get articles and letters 
    section = soup3.findAll('a', {'class':'fulltext'})

    l_a_section = []

    for a in section:
        # only select article and letter a tags
        if a['href'].startswith(url[:27] + 'full/nature'):
            l_a_section.append(a)
    
    # Get link for articles and letters
    l_a_link_list = pd.DataFrame(l.get('href') for l in l_a_section)
    l_a_link_list = l_a_link_list.rename(columns = {0:'Paper_link'})

    # Get Paper name in letters and articles
    l_a_names = [l.get('title') for l in l_a_section]

    # Get rid of all the crap in the name
    for x in xrange(len(l_a_names)):
#             l_a_names[x] = str(l_a_names[x])
            l_a_names[x] = l_a_names[x].replace('|[ggr]|',"g")
            l_a_names[x] = l_a_names[x].replace("'",'')
            l_a_names[x] = l_a_names[x].replace('|[thinsp]||[ap]||[thinsp]|','')
            l_a_names[x] = l_a_names[x].replace('|[ndash]|','-')
            l_a_names[x] = l_a_names[x].replace('|','')
            l_a_names[x] = l_a_names[x].replace("]", '')
            l_a_names[x] = l_a_names[x].replace("[", '')

    # Convert to dataframe
    l_a_names = pd.DataFrame(l_a_names)
    l_a_names = l_a_names.rename(columns = {0:'Paper_name'})

    # join names and links
    all_papers = pd.concat([l_a_names, l_a_link_list], axis = 1)
    all_papers['Paper_type'] = 'unknown'

    # Add paper volume and issue info
    all_papers['volume'] = url2[38:41]
    all_papers['issue'] = url2[43:47]

    # Add publication date
    pub_date_html = soup3.find("div", {"class": "issue-links"})
    pub_date_rev = pub_date_html('h3')
    pub_date = pub_date_rev[0].contents
    # Get rid of all the extra crap in the date
    pub_date = str(pub_date)
    pub_date = pub_date.replace("[u'In this issue (",'')
    pub_date = pub_date.replace(")']",'')      
    # Add to pub_date to df
    all_papers['pub_date'] = pub_date
    
    # Get rid of those pesky News and Views articles and references to other Articles/Letters
    all_papers = all_papers[~all_papers['Paper_name'].str.contains('News & Views')]
    all_papers = all_papers[~all_papers['Paper_name'].str.contains('et al.')]

    # Apend to master df
    sorted_all_papers = pd.concat([sorted_all_papers, all_papers], axis = 0, ignore_index = True)
```
